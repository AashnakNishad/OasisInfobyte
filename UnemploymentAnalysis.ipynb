**PROBLEM STATEMENT**-

Unemployment is measured by the unemployment rate which is the number of people
who are unemployed as a percentage of the total labour force. We have seen a sharp increase in the unemployment rate during Covid-19, so analyzing the unemployment rate can be a good data science project.

```python
#Importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the dataset

from google.colab import files
uploaded=files.upload()
```

```python
#Reading the dataset

df=pd.read_csv('Unemployment in India.csv')
df
```

```python
#renaming the columns

df.columns=['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate','Region']
```

```python
#Describing the datset

df.info()
df.describe()
df.head()
```

```python
#checking for missing values

df.isnull().sum()
```

```python
#Handling missing values

df.dropna(axis = 0, inplace = True)
df.isnull().sum()
```

```python
# area wise unemployment rate using pie chart

plt.figure(figsize=(10,5))
df.groupby('Region')['Estimated Unemployment Rate'].mean().plot(kind='pie', autopct='%1.1f%%')
plt.title('Area wise Unemployment Rate')
plt.show()
```

```python
# Heatmap

plt.figure(figsize=(10, 6))
corr = df[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']].corr()
sns.heatmap(corr, annot=True, linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
