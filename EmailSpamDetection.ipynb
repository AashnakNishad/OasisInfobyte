```python
#importing the libraries

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Importing the dataset

from google.colab import files
uploaded=files.upload()
```

```python
#Reading the dataset

df=pd.read_csv('spam.csv', encoding='latin1')
df
```

```python
#rename the columns

df = df.rename(columns={'v1': 'Category', 'v2': 'Message'})
df
```

```python
#checking the null values

df.isnull().sum()
```

```python
#replace the null values with a null string

df = df.where((pd.notnull(df)),'')

#Checking again
df.isnull().sum()
```

```python
#label encoding
#here we are converting the string data type to a binary number
#spam mail:0
#ham mail:1

df.loc[df['Category'] == 'spam', 'Category',] = 0
df.loc[df['Category'] == 'ham', 'Category',] = 1
```

```python
#separating the data sets as texts and label

X=df['Message']
Y=df['Category']
X
Y
```

```python
#splitting the data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
#transform the text data to feature  vectors that can be used as input to
#the logistic regression model

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english')
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
```

```python
#convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
#training the model

model=LogisticRegression()
model.fit(X_train_features,Y_train)
```

```python
#prediction on training data

prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
print('Accuracy on training data:',accuracy_on_training_data)
```

```python
#predicting on test data

prediction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
print('Accuracy on test data:',accuracy_on_test_data)
```

```python
#Building a predictve system

input_mail=["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]
#convert text to feature vectors
input_data_features=feature_extraction.transform(input_mail)
#making prediction
prediction=model.predict(input_data_features)
print(prediction)
if (prediction[0]==1):
  print('Ham mail')
else:
  print('Spam mail')
```
